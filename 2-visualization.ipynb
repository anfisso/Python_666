### 3 -  Ежемесячные данные о сне

# 4 - Занятость женщин и уровень рождаемости 

```python
fertility = pd.read_csv('df_/SLR2.csv')
fertility.head()
```

### - здесь мы видим очевидную картину, уровень рождаемости  растёт соответственно спаду занятости женщин 

```python
# создаем палитру 
bc = mcolors.XKCD_COLORS

# устанавливаем значения на осях 
x = fertility['fertility rate']
y = fertility['worker percent']

num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(x))
sizes = num_set * 35
colors = [list(bc.keys())[i] for i in num_set]

plt.scatter(x, y, s=sizes, alpha=0.4, c=colors, linewidths=2, edgecolors="face")
plt.plot(x, y, "g--", alpha=0.4)
plt.xlabel("fertility rate")
plt.ylabel("worker percent")
plt.grid(axis = 'y')
```
