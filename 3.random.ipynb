```python
import random
import numpy as np
```

```python
pay = 0            # платеж за игру, увеличиваем на каждой итерации (на 1)
winning = 0        # выигрыш 
replay = 1*10**6   # Количество игр 
```

```python

for i in range(replay):
    pay += 1               # на каждую итерацию тратим один рубль 
    rand = random.randrange(0, 1000)
    if rand == 777:
        winning += 200
    elif rand == 999:
        winning += 100
    elif rand == 555:
        winning += 50
    elif rand == 333:
        winning += 15
    elif rand == 111:
        winning += 10
    elif rand % 100 == 77:
        winning += 5
    elif rand % 10 == 7:
        winning += 3
    elif rand % 100 == int('00'):
        winning += 2
    elif rand % 10 == 0:
        winning += 1
    else:
        winning += 0
    
res = (winning / pay) - 1 

if winning <= pay:
    print(f'Игра не выгодна. Выигрыш {winning}руб., потрачено {pay}руб.')
else:
    print(f'Ваш доход составляет {winning-pay}руб.')
    
if res > 0:
    print(f'В среднем игрок за одну игру выигрывает {res}руб.')
else:
    print(f'В среднем игрок теряет за одну игру {-res}руб.')
```
